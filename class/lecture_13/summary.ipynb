```python
import matplotlib.pyplot as plt
import numpy as np

# Define the vectors
b1 = np.array([1, 2])
b2 = np.array([2, 0])
b_sum = b1 + b2  # Sum of b1 and b2

# Set up the plot
plt.figure()
plt.axhline(0, color='grey', lw=0.5)
plt.axvline(0, color='grey', lw=0.5)

# Plot b1
plt.quiver(0, 0, b1[0], b1[1], angles='xy', scale_units='xy', scale=1, color='blue', label='b1')
# Plot b2
plt.quiver(0, 0, b2[0], b2[1], angles='xy', scale_units='xy', scale=1, color='green', label='b2')
# Plot the resultant vector (b1 + b2)
plt.quiver(0, 0, b_sum[0], b_sum[1], angles='xy', scale_units='xy', scale=1, color='red', label='b1 + b2')

# Set plot limits
plt.xlim(0, 5)
plt.ylim(0, 5)

# Add grid, legend, and labels
plt.grid()
plt.legend()
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("2 space vectors")

# Show plot
plt.show()
```

### Length of vectors

```python
import numpy as np

def get_length(name, x, y):
  # Calculate the length (magnitude) of the vector
  return np.linalg.norm(np.array([x, y]))
```

```python
b1 = np.array([1, 2])
length_b1 = get_length("b1", b1[0], b1[1])
b2 = np.array([2, 0])
length_b2 = get_length("b2", b2[0], b2[1])
b3 = b1 + b2
length_b_sum = get_length("b1 + b2", b3[0], b3[1])

# Set up the plot
plt.figure()
plt.axhline(0, color='grey', lw=0.5)
plt.axvline(0, color='grey', lw=0.5)

# Plot b1
plt.quiver(0, 0, b1[0], b1[1], angles='xy', scale_units='xy', scale=1, color='blue', label='b1')
# Plot b2
plt.quiver(0, 0, b2[0], b2[1], angles='xy', scale_units='xy', scale=1, color='green', label='b2')
# Plot the resultant vector (b1 + b2)
plt.quiver(0, 0, b3[0], b3[1], angles='xy', scale_units='xy', scale=1, color='red', label='b1 + b2')

# Set plot limits
plt.xlim(0, 5)
plt.ylim(0, 5)

# Add annotations for vector lengths
plt.text(b1[0]/2, b1[1]/2, f"|b1| = {length_b1:.2f}", color="blue", ha='center')
plt.text(b2[0]/2, b2[1]/2, f"|b2| = {length_b2:.2f}", color="green", ha='center')
plt.text(b_sum[0]/2, b_sum[1]/2, f"|b1 + b2| = {length_b_sum:.2f}", color="red", ha='center')

# Add grid, legend, and labels
plt.grid()
plt.legend()
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("2 space vectors")

# Show plot
plt.show()
```

## Lattice

A lattice generated by two vectors \( b_1 \) and \( b_2 \) in \(\mathbb{R}^2\) can be defined as:

$$
\text{Lattice} = \{ n \cdot b_1 + m \cdot b_2 \mid n, m \in \mathbb{Z} \}
$$

where \( n \) and \( m \) are integers.

The span of two vectors \( b_1 \) and \( b_2 \) includes all linear combinations of these vectors:

$$
\text{Span}(b_1, b_2) = \{ \alpha \cdot b_1 + \beta \cdot b_2 \mid \alpha, \beta \in \mathbb{R} \}
$$

This defines the entire plane spanned by \( b_1 \) and \( b_2 \).


## Linearly Independent Vectors

The vector \( b_i \) is not a member of the span of \( b_2 \), which can be expressed as:

$$
b_i \notin \text{Span}(b_2)
$$

## Lattice: summary

### Lattice

A lattice \( \text{L} \) generated by a set of linearly independent vectors \( B = (b_1, b_2, b_3, \dots, b_n) \) can be written as:

$$
\text{L} = \{ x \cdot b \mid x \in \mathbb{Z} \}
$$

where each basis vector \( b_i \in \mathbb{Z}^m \) and satisfies the condition:

$$
b_i \notin \text{Span}(b_1, \dots, b_{i-1}, b_{i+1}, b_n)
$$

This ensures that each \( b_i \) is linearly independent from the other vectors in \( B \), forming a valid lattice basis.

if n >= 2 , there are infinitely many bases for L

- For n = 1,  it is the line
- For n = 2, it is the plane


## The Gram-Schmidt Orthogonalised (GSO) Basis

The **Gram-Schmidt process** is used to find a set of orthogonal vectors that are linearly independent and span the same vector space as the original set. Below are details about the process, how to evaluate lattice bases, and the distinction between good and bad bases.

---

### Steps of the Gram-Schmidt Process

1. **First Vector (\( b_1^* \))**:
   - Start with the first vector \( b_1 \) in the vector space.
   - The first orthogonal vector \( b_1^* \) is the same as \( b_1 \) because there are no preceding vectors to orthogonalize against.

2. **Second Vector (\( b_2^* \))**:
   - The second orthogonal vector \( b_2^* \) is computed by removing the projection of \( b_2 \) onto \( b_1^* \):
     \[
     b_2^* = b_2 - \text{proj}_{b_1^*}(b_2)
     \]
     where:
     \[
     \text{proj}_{b_1^*}(b_2) = \frac{\langle b_2, b_1^* \rangle}{\| b_1^* \|^2} b_1^*
     \]
   - This ensures \( b_2^* \) is orthogonal to \( b_1^* \), making the angle between them \( 90^\circ \).

3. **General Step (\( b_k^* \))**:
   - For any \( b_k^* \), remove projections onto all previously orthogonalized vectors \( b_1^*, b_2^*, \ldots, b_{k-1}^* \):
     \[
     b_k^* = b_k - \sum_{j=1}^{k-1} \text{proj}_{b_j^*}(b_k)
     \]
   - After this step, \( b_k^* \) is orthogonal to all \( b_1^*, b_2^*, \ldots, b_{k-1}^* \).

4. **Orthogonality Check**:
   - After orthogonalization:
     \[
     \langle b_i^*, b_j^* \rangle = 0 \quad \text{for } i \neq j
     \]

---

### Good vs. Bad Lattice Basis

#### Characteristics of a Good Basis
- **Short vectors** (small magnitudes).
- Vectors are **nearly orthogonal** (angles close to \( 90^\circ \)).

#### Characteristics of a Bad Basis
- **Long vectors** (large magnitudes).
- Vectors are **highly non-orthogonal** (angles far from \( 90^\circ \)).

#### Why It Matters
- In applications like **lattice-based cryptography** or computational problems (e.g., the Shortest Vector Problem), a "good" basis makes computation significantly easier.
- The **Gram-Schmidt orthogonalized basis** helps identify the quality of the original basis. Large variations in the magnitudes of \( b_k^* \) suggest the original basis is "bad."

---

### Example: Evaluating a Basis

1. **Start with a Basis**:
   \[
   \text{Original Basis: } B = \{b_1, b_2, b_3\}
   \]

2. **Compute the Orthogonalized Basis**:
   \[
   \text{GSO Basis: } B^* = \{b_1^*, b_2^*, b_3^*\}
   \]

3. **Evaluate**:
   - **Magnitudes** (\( \| b_k^* \| \)):
     - Large variations in the magnitudes of \( b_k^* \) indicate a "bad" basis.
   - **Angles**:
     - Check the angles between the original vectors. Angles far from \( 90^\circ \) indicate a "bad" basis.

### Improving a Bad Basis
- Use algorithms like the **LLL algorithm** (Lenstra-Lenstra-Lovsz) to reduce a "bad" basis.
- A reduced basis has shorter, nearly orthogonal vectors, making lattice problems easier to solve.

### Practical Applications
- Evaluating lattice bases is crucial in **lattice-based cryptography**.

  - **Shortest Vector Problem (SVP)**
  - **Closest Vector Problem (CVP)**

## Projection of \( b_k \) Orthogonal to the Span

The **Gram-Schmidt process** involves projecting a vector \( b_k \) onto the subspace spanned by the previous orthogonalized vectors \( b_1^*, b_2^*, \dots, b_{k-1}^* \), and then subtracting this projection to make \( b_k^* \) orthogonal to that span.

---

### Mathematical Formula

For a vector \( b_k \), the orthogonal projection onto the span of \( b_1^*, b_2^*, \dots, b_{k-1}^* \) is given by:

\[
\pi_{\text{span}}(b_k) = \sum_{j=1}^{k-1} \text{proj}_{b_j^*}(b_k)
\]

where:

\[
\text{proj}_{b_j^*}(b_k) = \frac{\langle b_k, b_j^* \rangle}{\| b_j^* \|^2} b_j^*
\]

The orthogonal component of \( b_k \) (denoted \( b_k^* \)) is computed as:

\[
b_k^* = b_k - \pi_{\text{span}}(b_k)
\]

---

### Steps to Compute \( b_k^* \)

1. **Start with \( b_k \)**:
   - Begin with the \( k \)-th vector from the original basis.

2. **Project \( b_k \) onto each \( b_j^* \) (\( j < k \))**:
   - Calculate the projection of \( b_k \) onto \( b_j^* \) using:
     \[
     \text{proj}_{b_j^*}(b_k) = \frac{\langle b_k, b_j^* \rangle}{\| b_j^* \|^2} b_j^*
     \]

3. **Sum All Projections**:
   - Sum the projections to find \( \pi_{\text{span}}(b_k) \):
     \[
     \pi_{\text{span}}(b_k) = \sum_{j=1}^{k-1} \text{proj}_{b_j^*}(b_k)
     \]

4. **Compute \( b_k^* \)**:
   - Subtract the projection from \( b_k \) to get the orthogonal component:
     \[
     b_k^* = b_k - \pi_{\text{span}}(b_k)
     \]

---

### Example

Given the basis:
\[
B = \{b_1, b_2, b_3\}
\]

And the already orthogonalized vectors:
\[
B^* = \{b_1^*, b_2^*\}
\]

To compute \( b_3^* \):

1. Compute projections:
   \[
   \text{proj}_{b_1^*}(b_3) = \frac{\langle b_3, b_1^* \rangle}{\| b_1^* \|^2} b_1^*
   \]
   \[
   \text{proj}_{b_2^*}(b_3) = \frac{\langle b_3, b_2^* \rangle}{\| b_2^* \|^2} b_2^*
   \]

2. Find \( \pi_{\text{span}}(b_3) \):
   \[
   \pi_{\text{span}}(b_3) = \text{proj}_{b_1^*}(b_3) + \text{proj}_{b_2^*}(b_3)
   \]

3. Subtract to get \( b_3^* \):
   \[
   b_3^* = b_3 - \pi_{\text{span}}(b_3)
   \]

---

### Summary

- The orthogonal component \( b_k^* \) is found by removing the part of \( b_k \) that lies in the span of \( \{b_1^*, b_2^*, \dots, b_{k-1}^*\} \).
- This ensures that \( b_k^* \) is orthogonal to all previous vectors in the orthogonalized basis.

---

### Applications

- **Lattice Basis Reduction**: Identifying orthogonal components is essential in improving lattice bases for cryptography and computational geometry.
- **Numerical Stability**: Ensures better numerical behavior in algorithms requiring orthogonality, like QR decomposition.
